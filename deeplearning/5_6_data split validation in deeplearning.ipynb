```python
import numpy as np
import math
import matplotlib.pyplot as plt
import tensorflow as tf
import pandas as pd
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('~/dropbox/python숙달/data/pima-indians-diabetes.data.csv')
X = df.iloc[:,:-1].values
y = df.iloc[:, -1:].values
```

# scaling

```python
scaler = MinMaxScaler()
x_scalerd = scaler.fit_transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scalerd, y, test_size=0.3, stratify= y)
```

# valid data를 두어 accuracy를 확인하는 것도 가능.

```python
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.2, stratify= y_train)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
dense = Dense( units = 1, input_dim = 8, activation = 'sigmoid')
model = Sequential([dense])
model.compile(loss = "binary_crossentropy", optimizer = Adam(0.1), metrics = ['acc'])
h = model.fit(x_train, y_train, epochs= 1000, validation_data= (x_valid, y_valid))
```

```python
## history fit 내부에서 해주는 것도 가능.
history = model.fit(x_train, y_train, epochs=15, batch_size=64, validation_split=0.2)
```

```python
h.history.keys()
```

```python
plt.plot(h.history['loss'])
plt.plot(h.history['val_loss'])
```

```python
plt.plot(h.history['acc'])
plt.plot(h.history['val_acc'])
```

```python
model.evaluate(x_train, y_train)
```

```python
model.evaluate(x_test, y_test)
```
