# sigmoid function 이해하기

```python
def sigmoid( z ):
    return 1 / (1+ math.e ** -z)
```

```python
print(sigmoid(-10))
print(sigmoid(-1))
print(sigmoid(0))
print(sigmoid(1))
print(sigmoid(10))
```

```python
for z in np.linspace(-10, 10, 50):
    s = sigmoid( z )
    plt.plot( z, s, 'ro')
plt.show()
```

## keras with pipeline

```python
def createModel():
    dense = Dense(units = 1, input_dim = 8, activation = 'sigmoid')
    model = Sequential([dense])
    model.compile(loss = "binary_crossentropy", optimizer = Adam(0.1), metrics = ['acc'])
    return model
```

```python
earlyStop = EarlyStopping(monitor = 'loss', min_delta = 0.01, patience = 100)
kcf = KerasClassifier(build_fn = createModel, epochs = 1000, verbose = 1, callbacks = [earlyStop])
```

```python
## pipeline을 통과시키면 자동으로 predict도 fit을 해준다. 
pmodel = make_pipeline( MinMaxScaler(), kcf)
pmodel.fit(x, y)
```

```python
pmodel.predict( [[6,148,72,35,0,33.6,0.627,50]] )
```
