# Cost function using TF

```python
import tensorflow as tf
from tensorflow.keras.optimizers import SGD
```

```python
x_data = [1,2,3]
y_data = [1,2,3]
```

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
w = 3
```

```python
hx = w * x # tf.multiply( w, x)
sq = tf.square(hx - y) #( hx-y) ** 2
cost = tf.reduce_mean( sq)
```

```python
def cost(x, y, w):
    hx = w * x
    cost = tf.reduce_mean(tf.square(hx - y))
    return cost
```

## history of optimizer

- SGD: 확률을 이용해서 모든 step을 다 보지 않고, 수렴시킴.
- momentum: 관성의 법직을 이용. 기존 가는 방향에 가중치를 부여.
- 

일반적으로는 optimizer를 ADAM이 goodness

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
## 값이 계속 바뀌는 부분은 항상 constant가 아니라 variable을 찾는다

w = tf.Variable(10.0)
b = tf.Variable(10.0)
```

```python
def compute_cost():
    hx = w * x + b
    cost = tf.reduce_mean(( hx - y) ** 2)
    return cost
```

```python
optimizer = SGD(learning_rate = 0.01) # 미분 + a
```

```python
for i in range(2000):
    optimizer.minimize(compute_cost, var_list = [w, b])
    # w = w- 0.01 * w에 대한 미분값.w.assign_sub(0.01 * 미분값) 미분한 값으로 계속 주기적으로 바꿔주는 것.
    # b = b -0*x에 대한 미분값. b.assign_sub(0.01 * 미분값) 미분한 값으로 계속 주기적으로 바꿔주는 것.
    print(i, "cost", compute_cost().numpy())
```
