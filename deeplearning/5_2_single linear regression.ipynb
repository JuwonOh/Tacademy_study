# TF를 사용해서 single linear regression 구현.

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
w = tf.Variable(tf.random.uniform([1]))
b = tf.Variable(tf.random.uniform([1]))
# init을 어떻게 주는가에 대한 알고리즘이 savior가 있음.
```

```python
def compute_cost():
    hx = w * x + b
    cost = tf.reduce_mean(( hx - y) ** 2)
    return cost
```

```python
optimizer = Adam(learning_rate= 0.01)
```

```python
list_cost = []
for i in range(2000):
    
    optimizer.minimize(compute_cost, var_list = [w, b])
    c = compute_cost().numpy()
    list_cost.append(c)
    print(i , "cost = ", c)
```

```python
dense = Dense( units = 1, input_dim =1 ) 
#weight와 bias가 들어가 있는 dense를 만든다. unit = # label(dist 제동거리), input dim = # feature(speed) 최적의 inir값을 구해줌.
model = Sequential([dense])
model.compile(loss='mse', optimizer = Adam(learning_rate=0.01)) # cost function인 mse를 구현하지 않아도 기본적인 틀이 완성되어 있음.
h = model.fit(x_data, y_data, epochs = 1000)
```

# torch로 해보기

```python
import torch
from torch.optim import Adam
```

```python
x_data = torch.FloatTensor([1,2,3,4,5])
y_data = torch.FloatTensor([6,7,8,9,10])
```

```python
w = torch.tensor(10.0, requires_grad = True)# True면 Variable 변수가 된다.
b = torch.tensor(10.0, requires_grad = True)
```

```python
optimizer = Adam([w, b], lr = 0.1)
```

```python
for i in range(2000):
    hx = w * x_data + b
    cost = torch.mean((hx - y_data)**2)
    optimizer.zero_grad()# grad 값을 zeos해줌.
    cost.backward() # 미분 함수. w, b에 대한 미분 수행
    optimizer.step() # w 미분값, b에 대한 미분 값
    print(i, 'step cost = ', cost.item())
```

## mse(결정계수?? r^2)로 검증하시오.

- mse가 cost function이라서 2999 step cost =  0.02514859102666378

$R^{2} = 1 - \frac{\sum_{i}(y_i-\hat{y_i})^2}{\sum_i(y_i-\overline{y})^2}$

```python
y_pred = hxfn(x_data)
```

```python
def r2_loss(output, target):
    target_mean = torch.mean(target)
    ss_tot = torch.sum((target - target_mean) ** 2)
    ss_res = torch.sum((target - output) ** 2)
    r2 = 1 - ss_res / ss_tot
    return r2
```

```python
r2_loss(y_pred, y_data)
```
