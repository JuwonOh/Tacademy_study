```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
```

# multiple regression

```python
x1_data = [1,0,3,4,5] # 공부 시간
x2_data = [0,1,2,3,4] # 출석 시간
y_data = [1,2,3,4,5] # 점수
```

```python
x1 = tf.constant(x1_data, tf.float32)
x2 = tf.constant(x2_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
w1 = tf.Variable(tf.random.uniform([1]))
w2 = tf.Variable(tf.random.uniform([1]))
b = tf.Variable(tf.random.uniform([1]))
```

```python
def compute_loss():
    hx = w1 * x1 + w2 * x2 + b
    cost = tf.reduce_mean((hx - y)**2)
    return cost
```

```python
optimizer = Adam(0.1)
```

```python
for i in range(2000):
    optimizer.minimize(compute_loss, var_list = [w1, w2, b])
    print(i, "cost = ", compute_loss().numpy())
```

```python
w1.numpy(), w2.numpy(), b.numpy()
```

```python
#5시간공부 5시간 출석
w1.numpy()[0]*5 + w1.numpy()[0]*5 + b.numpy()[0]
```

# toy data

```python
df = pd.read_csv('~/dropbox/python숙달/data/depp/test.csv', header = None)
df.columns =['Q1', 'Q2','midterm', 'final']
```

```python
df
```

```python
x_data = df.iloc[:,:-1].values
y_data = df.iloc[:,-1:].values
```

```python
x_data.shape
```

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
w = tf.Variable(tf.random.uniform([3,1])) # [특성데이터, 라벨의 수]
b = tf.Variable(tf.random.uniform([1]))
```

```python
def compute_loss():
    hx = tf.matmul(x, w) + b
    cost = tf.reduce_mean((hx - y) **2)
    return cost
```

```python
compute_loss
```

```python
optimizer = Adam(0.1)
for i in range(3000):
    optimizer.minimize(compute_loss, var_list = [w, b])
    print(i, "cost =", compute_loss().numpy())
```

```python
def hxfn(xdata):
    xdata = tf.constant(xdata, tf.float32)
    hx = tf.matmul(xdata, w) + b
    return hx.numpy()
```

```python
plt.plot( y_data)
plt.plot( hxfn( x_data) ,'r--')
plt.show()
```
