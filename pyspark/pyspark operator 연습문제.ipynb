```python
from pyspark import SparkContext
```

```python
sc = SparkContext('local')
```

```python
rdd1 = sc.textFile('data/k.txt')
```

# k.txt 에 있는 각 단어의 갯수를 구하시요

```python
rdd1.flatMap(lambda x: x.split(' ')).map(lambda v: (v,1)).countByKey()
```

# births.txt 파일을 이용하여 구하시요

1) 남아수의 합과 평균을 구하시요

```python
rdd3 = sc.textFile('data/births.txt')
```

```python
rdd3.map(lambda v: float(v.split(',')[1])).sum()
```

```python
rdd3.map(lambda v: float(v.split(',')[1])).mean()
```

여아수의 합과 평균을 구하시요 

```python
rdd3.map(lambda v: float(v.split(',')[2])).sum()
```

```python
rdd3.map(lambda v: float(v.split(',')[2])).mean()
```

3) 남아수가 높은 탑 5를 구하시요 

```python
rdd3.sortBy(lambda v: float(v.split(',')[1]), ascending=False).take(5)
```

4) 전체(남아여아의총수)중 남아수의 비율을 구하시요.

```python
rdd3.map(lambda v: float(v.split(',')[1]) / (float(v.split(',')[1]) + float(v.split(',')[2]))).collect()
```

 5) 여아수를 10등분 했을때 각계급의 수를 구하시요

```python
rdd3.map(lambda v:v.split(',')).map(lambda v:float(v[2])).histogram(10)[1]
```
